## Customizing Legends

````python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
np.random.seed(0)
````

````python
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
ax.legend();
````

````python
ax.legend(frameon=False, loc='upper left')
fig
````

````python
ax.legend(frameon=False, loc='lower center')
fig
````

````python
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))

plt.legend(['first', 'second']);
````

````python
plt.plot(x, np.cos(x), label="first")
plt.plot(x, np.sin(x), label="second")

plt.legend();
````

## Customizing Color Bars

````python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])
````

````python
plt.imshow(I, cmap='gray');
````

````python
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(
    loc=0.0,
    scale=3.0,
    size=np.count_nonzero(speckles)
)

plt.figure(figsize=(10, 3.5))

plt.subplot(1, 2, 2)
plt.imshow(I, cmap='RdBu')
plt.colorbar(extend='both')
plt.clim(-1, 1)
````

## Add text to figure

```python
plt.text(
    xpos,
    ypos,
    text,
    horizontalalignment='center',
    verticalalignment='center',
    ...
)
```

````python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x), label="first")
ax.plot(x, np.sin(x), label="second")
ax.legend();

fig.suptitle(
    'bold figure suptitle',
    fontsize=14,
    fontweight='bold'
)

ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

ax.text(
    2, 2,
    'boxed italics text in data coords',
    bbox={'facecolor': 'red', 'alpha': 0.5}
)

ax.plot([np.pi / 2.0], [1], 'o')
ax.annotate(
    '$cos(\pi / 2)$',
    xy=(np.pi / 2.0, 1),
    xytext=(np.pi / 2.0 - 0.65, 1.75),
    arrowprops=dict(facecolor='black', shrink=0.05)
)

plt.show();
````
